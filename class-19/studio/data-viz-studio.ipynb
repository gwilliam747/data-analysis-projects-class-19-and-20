```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
print(df["year"].unique())
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
# Tried filtering by year and found that the column is strings instead of int.  Converting:
df["year"].dtype
df["year"] = pd.to_numeric(df["year"])
viz1 = df[(df["year"] > 1999) & (df["year"] < 2010)]

books_published_1 = viz1.groupby("year").size()


plt.bar(books_published_1.index, books_published_1.values)
plt.title("Books Published per Year")
plt.xlabel("Year")
plt.ylabel("Books")
plt.xticks(books_published_1.index)
plt.show()
```

```python
# Use your Python skills to create a visualization here.
viz2 = df[(df["year"] > 2009) & (df["year"] < 2020)]
books_published_2 = viz2.groupby("year").size()

plt.bar(books_published_2.index, books_published_2.values)
plt.title("Books Published per Year")
plt.xlabel("Year")
plt.ylabel("Books")
plt.xticks(books_published_2.index)
plt.show()
```

According to this data, there was a dramatic publication dropoff in 2007, after steady increase before then.  The 2010s continued this trend, with 10 or fewer books being published each year after 2013
